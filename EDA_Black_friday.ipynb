```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_final.drop(columns=['User_ID'],axis=1,inplace = True)
```

```python
df_final
```

```python
pd.get_dummies(df_final['Gender'])
```

```python
# Handling categorical featue Gender
df_final['Gender'] = df_final['Gender'].map({'F':0,'M':1}) #Converted into only in one line of code
```

```python
df_final
```

```python
# Handling categorical fetaure Age
df_final['Age'].unique()
```

```python
pd.get_dummies(df_final['Age'],drop_first=True,dtype=int)
```

```python
df_final['Age'] = df_final['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df_final.sample(14)
```

```python
# Handling City_Category
df_city = pd.get_dummies(df_final['City_Category'],dtype = int,drop_first = True)
```

```python
df_city.head()
```

```python
df_final = pd.concat([df_final,df_city],axis = 1)
```

```python
df_final.drop(columns=['City_Category'],inplace = True,axis = 1)
```

```python
df_final
```

```python
#Mising value
df_final.isnull().sum()
```

```python
# Focus on replacing missing value
df_final.Product_Category_2.unique()
```

```python
df_final.Product_Category_2.value_counts()
```

```python
# replkace the missing value with mode  
df_final['Product_Category_2'] = df_final['Product_Category_2'].fillna(df_final['Product_Category_2'].mode()[0])
```

```python
df_final
```

```python
df_final['Product_Category_3'].unique()
```

```python
df_final['Product_Category_3']  =df_final['Product_Category_3'].fillna(df_final['Product_Category_3'].mode()[0])
```

```python
df_final.isnull().sum()
```

```python
# stay_in_current_city
df_final['Stay_In_Current_City_Years'].unique()
```

```python
# df_final['Stay_In_Current_City_Years'] = df_final['Stay_In_Current_City_Years'].map({'0':0,'1':1,'2':2,'3':3,'4+':4})
df_final['Stay_In_Current_City_Years'] = df_final['Stay_In_Current_City_Years'].str.replace('+',"")
```

```python
df_final['Stay_In_Current_City_Years'].isnull().sum()
```

```python
# Convert object into integer
df_final['Stay_In_Current_City_Years'] = df_final['Stay_In_Current_City_Years'].astype(int)
```
